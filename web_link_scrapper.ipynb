```python
import requests 
import bs4
```

```python
res=requests.get("http://villagemap.in/andaman-and-nicobar-islands.html")
page=res.content
soup=bs4.BeautifulSoup(page,'html.parser')
texts=soup.find_all("div",id="m-content")
row_data=soup.findAll('tr')
```

```python
table_rows1=[]
first_page_links=[]
for row in range(1,len(row_data)):
    row1=row_data[row]
    current_row=[]
    rows=row1.findAll('td',{})
    link=row1.find_all('a',)
    l=str(link)
    l=l.split(" ")
    first_page_links.append(l[1][6:-15])
    
    for data in rows:
        current_row.append(data.text)
    table_rows1.append(current_row)
    
        
```

```python
for l in range(len(first_page_links)):
    first_page_links[l]='http://villagemap.in'+first_page_links[l]
    print(first_page_links[l])
```

```python
table_rows2=[]
second_page_links=[]
for li in first_page_links:
    res=requests.get(li)
    page=res.content
    soup=bs4.BeautifulSoup(page,'html.parser')
    texts=soup.find_all("div",id="m-content")
    row_data=soup.findAll('tr')
    for row in range(1,len(row_data)):
        row=row_data[row]
        current_row=[]
        rows=row.findAll('td',{})
        link=row.find_all('a')
        l=str(link)
        l=l.split(" ")
        second_page_links.append(l[1][6:-1])
        for idx,data in enumerate(rows):
            current_row.append(data.text)
        table_rows2.append(current_row)
        
```

```python
second_linksp = []
for l in second_page_links:
    l = l.split('"')
    second_linksp.append(l[0])
```

```python
for l in range(len(second_linksp)):
    
    second_page_links[l]='http://villagemap.in'+second_linksp[l]
    print(second_page_links[l])
```

```python
current_row = []
table_row3 = []
third_page_links = []
for li in second_page_links:
    res=requests.get(li)
    page=res.content
    soup=bs4.BeautifulSoup(page,'html.parser')
    texts=soup.find_all("div",id="m-content")
    link2 = soup.find_all('li')
    for i in range(len(link2)):
        l = link2[i].find_all('a',{})
        l=str(l)
        l=l.split(" ")
        l = l[1][6:]
        lin = l.split('"')
        
        link3 = lin[0]
        
        tex = lin[1][1:]
       
        third_page_links.append(link3)
       
        table_row3.append(tex)
    
        
```

```python
for l in range(len(third_page_links)):
    third_page_links[l]='http://villagemap.in'+third_page_links[l]
    print(third_page_links[l])
print(len(third_page_links))
```
